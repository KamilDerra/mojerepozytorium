                                    Zadanie 3: Budowa modelu sieci neuronowej.

                           Zbuduj model sieci neuronowej w PyTorch do klasyfikacji obrazów.
                           

Sieci neuronowe składają się z warstw/modułów wykonujących operacje na danych. Przestrzeń nazw torch.nn zapewnia wszystkie elementy potrzebne do zbudowania własnej sieci neuronowej. Każdy moduł w PyTorch jest podklasą nn.Module. Sieć neuronowa sama w sobie jest modułem składającym się z innych modułów (warstw). Ta zagnieżdżona struktura umożliwia łatwe budowanie złożonych architektur i zarządzanie nimi.

```python
import os
import torch
from torch import nn
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
```

Definiujemy naszą sieć neuronową poprzez podklasę nn.Module i inicjujemy warstwy sieci neuronowej w init. Każda podklasa nn.Module implementuje operacje na danych wejściowych w metodzie forward.

```python
class NeuralNetwork(nn.Module):                           # klasa NeuralNetwork dziedziczy po nn.Module
    def __init__(self):                                   # metoda __init__ inicjuje wartswy modelu
        super().__init__()                                # wywołanie konstruktora klasy bazowej nn.Module
        self.flatten = nn.Flatten()         # tworzymy obiekt warstwwy Flatten, który przekształca tensor wejściowy na tensor 1D
 # tworzymy obiekt sekwencyjny 'Sequential', który składa się z 3 warstw liniowych, z których każda jest poprzedzona funkcją
 # aktywacji ReLU:
        self.linear_relu_stack = nn.Sequential(
            nn.Linear(28*28, 512),                     # 1 wartstwa przyjmuje wejście o rozm. 28x28 i zwraca tensor o rozm. 512
            nn.ReLU(),                                 # funkcja aktywacji ReLU
            nn.Linear(512, 512),   # druga warstwa przyjmuje wyjście z pierwszej (jako wejście) i zwraca inny tensor o rozm. 512
            nn.ReLU(),
            nn.Linear(512, 10),  # trzecia warstwa przyjmuje wyjście z drugiej warstwy (jako wejście) i zwraca tensor o rozm. 10 
        )

    def forward(self, x):                             # metoda 'forward' (dla klasy 'self') definiuje przepływ modelu 'w przód' 
        x = self.flatten(x)                           # metoda 'flatten(x)' spłaszcza wejściowy tensor x
        logits = self.linear_relu_stack(x)            # metoda 'linear_relu_stack(x)' zwraca wynik operacji liniowej na x,
                                                      # z aktywacją ReLU
        return logits
```

```python
# Tworzymy instancję NeuralNetwork i drukujemy jej strukturę:

model = NeuralNetwork()
print(model)
```

Aby skorzystać z modelu przekazujemy mu dane wejściowe. Spowoduje to wykonanie modelu (metoda 'w przód') wraz z pewnymi operacjami w tle.

Wywołanie modelu na wejściu zwraca dwuwymiarowy tensor, w którym dim=0 odpowiada każdemu wyjściu zawierającemu 10 nieprzetworzonych wartości przewidywanych dla każdej klasy, a dim=1 odpowiada poszczególnym wartościom każdego wyjścia. Prawdopodobieństwa przewidywania uzyskujemy 'przepuszczając' je przez instancję modułu nn.Softmax:

```python
X = torch.rand(1, 28, 28)
X = train_features
logits = model(X)
pred_probab = nn.Softmax(dim=1)(logits)
y_pred = pred_probab.argmax(1)
print(f"Predicted class: {y_pred}")
```

```python
input_image = X
print(input_image.size())
```

```python
# Inicjujemy warstwę nn.Flatten, aby przekonwertować każdy obraz 2D 28x28 na ciągłą tablicę wartości 784 pikseli 
# (zachowany zostaje wymiar minibatch (przy dim=0):

flatten = nn.Flatten()
flat_image = flatten(input_image)
print(flat_image.size())
```

```python
# Warstwa liniowa (nn.Linear) to moduł, który stosuje transformację liniową na wejściu przy użyciu przechowywanych 
# wag i odchyleń:

layer1 = nn.Linear(in_features=28*28, out_features=64)
hidden1 = layer1(flat_image)
print(hidden1.size())
```

```python
# Aktywacje nieliniowe (nn.ReLU) tworzą złożone odwzorowania pomiędzy wejściami i wyjściami modelu. Stosuje się je PO (!) 
# przekształceniach liniowych w celu wprowadzenia nieliniowości, pomagając sieciom neuronowym uczyć się wielu różnych zjawisk.
# W tym modelu używamy nn.ReLU pomiędzy naszymi warstwami liniowymi:

print(f"Before nn.ReLU: {hidden1}\n\n")
hidden1 = nn.ReLU()(hidden1)
print(f"After nn.ReLU: {hidden1}")
```

```python
# 'nn.Sequential' to uporządkowany kontener modułów. Dane są przekazywane przez wszystkie moduły w tej samej kolejności, 
# jaka została zdefiniowana. Można użyć kontenerów sekwencyjnych, aby stworzyć szybką sieć, taką jak seq_modules:

seq_modules = nn.Sequential(
    flatten,
    layer1,
    nn.ReLU(),
    nn.Linear(64, 64)
)

logits = seq_modules(flat_image)
logits
```

```python
# Ostatnia warstwa liniowa sieci neuronowej zwraca logity (powyżej) – surowe wartości w [-infty, infty] – które przekazywane są 
# do modułu nn.Softmax. Logity są skalowane do wartości [0, 1] reprezentujących prawdopodobieństwa przewidywane przez model 
# dla każdej klasy. dim parametr wskazuje wymiar, wzdłuż którego wartości muszą sumować się do 1:

softmax = nn.Softmax(dim=1)
pred_probab = softmax(logits)
pred_probab
```

```python
# Parametry modelu.

# Wiele warstw w sieci neuronowej jest sparametryzowanych, tj. mają powiązane wagi i odchylenia, które są optymalizowane
# podczas uczenia. Podklasa nn.Module automatycznie śledzi wszystkie pola zdefiniowane w obiekcie modelu i udostępnia
# wszystkie parametry za pomocą metod naszego modelu: 'parameters()' lub 'named_parameters()'.

# W tym przykładzie iterujemy po każdym parametrze i drukujemy jego rozmiar oraz podgląd jego wartości:

print(f"Model structure: {model}\n\n")

for name, param in model.named_parameters():
    print(f"Layer: {name} | Size: {param.size()} | Values : {param[:2]} \n")
```

                                            OPTYMALIZACJA PARAMETRÓW MODELU

Teraz, gdy mamy model i dane, czas na uczenie, walidację i testowanie naszego modelu poprzez optymalizację jego parametrów na naszych danych. Uczenie modelu jest procesem iteracyjnym; w każdej iteracji model zgaduje wynik, oblicza błąd w swoim przypuszczeniu (stratę), zbiera pochodne błędu względem jego parametrów i optymalizuje te parametry za pomocą opadania gradientu.

                                                      Hiperparametry

Hiperparametry to regulowane parametry, które pozwalają kontrolować proces optymalizacji modelu. Różne wartości hiperparametrów mogą mieć wpływ na uczenie modelu i współczynniki zbieżności.

Definiujemy następujące hiperparametry do treningu:

Liczba epok — liczba iteracji po zbiorze danych

Batch Size – liczba próbek danych przesłanych w sieci przed aktualizacją parametrów

Szybkość uczenia się — stopień aktualizacji parametrów modeli w każdej partii/epoce. Mniejsze wartości powodują powolne tempo uczenia się, natomiast duże wartości mogą powodować nieprzewidywalne zachowanie podczas treningu.

```python
learning_rate = 1e-3
batch_size = 64
epochs = 5
```

                                                    Pętla optymalizacji

Po ustawieniu hiperparametrów możemy trenować i optymalizować nasz model za pomocą pętli optymalizacyjnej. Każda iteracja pętli optymalizacyjnej nazywana jest epoką.

Każda epoka składa się z dwóch głównych części:

Pętla pociągowa — iterujemy po zbiorze danych szkoleniowych i próbujemy uzyskać zbieżność do optymalnych parametrów.

Pętla walidacyjna/testowa — iterujemy po testowym zestawie danych, aby sprawdzić, czy wydajność modelu poprawia się.

                                                       Funkcja straty

Nasza nieprzeszkolona sieć prawdopodobnie nie udzieli poprawnej odpowiedzi, gdy przedstawimy ją z pewnymi danymi szkoleniowymi. Funkcja straty mierzy stopień odmienności uzyskanego wyniku od wartości docelowej i to właśnie tę funkcję straty chcemy minimalizować podczas treningu. Aby obliczyć stratę, dokonujemy prognozy na podstawie danych wejściowych z danej próbki danych i porównujemy ją z prawdziwą wartością etykiety danych.

Typowe funkcje straty obejmują nn.MSELoss (błąd średniokwadratowy) do zadań regresji i nn.NLLLoss (logarytm wiarygodności ujemnej) do klasyfikacji. nn.CrossEntropyLoss łączy nn.LogSoftmax i nn.NLLLoss.

Istnieją jeszcze np.: nn.L1Loss(), nn.BCELoss().

Przekazujemy logity wyjściowe naszego modelu do nn.CrossEntropyLoss, który normalizuje logity i oblicza błąd przewidywania.

                                                      Optymalizator

Optymalizacja to proces dostosowywania parametrów modelu w celu zmniejszenia błędów modelu na każdym etapie uczenia. Algorytmy optymalizacyjne definiują sposób realizacji tego procesu (w tym przykładzie używamy Stochastic Gradient Descent). Cała logika optymalizacji jest zawarta w obiekcie optymalizatora. Tutaj używamy optymalizatora SGD; dodatkowo w PyTorch dostępnych jest wiele różnych optymalizatorów, takich jak ADAM i RMSProp, Adagrad, Adadelta czy LBFGS, które działają lepiej w przypadku różnych rodzajów modeli i danych.

Inicjujemy optymalizator, rejestrując parametry modelu, które wymagają uczenia, i przekazując hiperparametr szybkości uczenia się.

```python
optimizer = torch.optim.SGD(model.parameters(), lr = learning_rate)
optimizer
```

W pętli treningowej optymalizacja przebiega w trzech etapach:

Wywołujemy optymalizator.zero_grad(), aby zresetować gradienty parametrów modelu. Gradienty domyślnie się sumują; aby zapobiec podwójnemu liczeniu, jawnie zerujemy je przy każdej iteracji.

Propagujemy wstecz prognozowaną stratę, wywołując metodę loss.backward(). PyTorch osadza gradienty straty w.r.t. każdy parametr.

Gdy już mamy gradienty, wywołujemy metodę optymalizacji.step(), aby dostosować parametry na podstawie gradientów zebranych w przebiegu wstecznym.

Definiujemy train_loop, który zapętla nasz kod optymalizacyjny, oraz test_loop, który ocenia wydajność modelu w porównaniu z naszymi danymi testowymi.

```python
train_dataloader == test_dataloader
```

```python
train_loss_history = []
test_loss_history = []
train_accuracy_history = []
test_accuracy_history = []

def train_loop(dataloader, model, loss_fn, optimizer):
    size = len(dataloader.dataset)
    
    # Ustawiamy model na tryb uczenia – ważne dla normalizacji wsadowej i warstw porzucania
    # Niepotrzebne w tej sytuacji, ale dodane w celu uzyskania najlepszych praktyk
    
    model.train()
    train_loss = 0
    train_accuracy = 0
    correct = 0
    
    def accuracy_fn(y, pred):
        correct = torch.eq(y, pred).sum().item()
        accuracy = (correct/len(pred))*100
        return accuracy
    
    for batch, (X, y) in enumerate(dataloader):
        
        # Obliczamy przewidywania i straty
        
        pred = model(X)
        loss = loss_fn(pred, y)
         
        
        # Backpropagation
        
        loss.backward()
        optimizer.step()
        optimizer.zero_grad()
        train_loss += loss
        train_accuracy += accuracy_fn(y, pred.argmax(dim=1))
        
    if batch % 100 == 0:
        loss, current = loss.item(), (batch + 1) * len(X)
        print(f"loss: {loss:>7f}  [{current:>5d}/{size:>5d}]")
        
    train_loss /= len(dataloader)
    train_accuracy /= len(dataloader)
      
    print(f"Train Error: \n Accuracy: {(train_accuracy):>0.1f}%, Avg loss: {train_loss:>8f} \n")
    train_accuracy_history.append(train_accuracy)
    train_loss_history.append(train_loss)
            
            

def test_loop(dataloader, model, loss_fn):
    
    # Ustawiamy model w tryb ewaluacyjny – ważny dla normalizacji wsadowej i warstw porzucania
    # Niepotrzebne w tej sytuacji, ale dodane w celu uzyskania najlepszych praktyk
    
    model.eval()
    size = len(dataloader.dataset)
    num_batches = len(dataloader)
    test_loss, test_accuracy = 0, 0
  
    
    # Ocena modelu za pomocą torch.no_grad() zapewnia, że w trybie testowym nie są obliczane żadne gradienty
    # Służy również do zmniejszenia niepotrzebnych obliczeń gradientu i zużycia pamięci dla tensorów za pomocą require_grad=True
    
    with torch.no_grad():
        for X, y in dataloader:
            pred = model(X)
            test_loss += loss_fn(pred, y).item()
            test_accuracy += (pred.argmax(1) == y).type(torch.float).sum().item()



    test_loss /= num_batches
    test_accuracy /= size
    print(f"Test Error: \n Accuracy: {(100*test_accuracy):>0.1f}%, Avg loss: {test_loss:>8f} \n")
    test_accuracy_history.append(100*test_accuracy)
    test_loss_history.append(test_loss)
    
    
```

```python
# Inicjujemy funkcję straty i optymalizator, a następnie przekazujemy je do train_loop i test_loop. 
# Możemy zwiększyć liczbę epok, aby śledzić poprawę wydajności modelu:

loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)

epochs = 10

for t in range(epochs):
    print(f"Epoch {t+1}\n-------------------------------")
    train_loop(train_dataloader, model, loss_fn, optimizer)
    test_loop(test_dataloader, model, loss_fn)            
print("Koniec pętli!")
```
