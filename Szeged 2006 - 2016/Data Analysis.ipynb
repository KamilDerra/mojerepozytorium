```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
df = pd.read_csv('Dane/weatherHistory.csv')
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
df.isnull().sum()
```

```python
df['Formatted Date'].dtype
```

```python
df['Formatted Date'] = df['Formatted Date'].str.split(" ").map(lambda x: x[0])
```

```python
df.head()
```

```python
df = df.drop('Daily Summary', axis=1)
df.head()
```

```python
df['Formatted Date'] = pd.to_datetime(df['Formatted Date'], errors='coerce')
df['Formatted Date'] = df['Formatted Date'].dt.year
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Summary'].value_counts()
```

```python
df['Precip Type'].value_counts()
```

```python
df['Loud Cover'].value_counts()
```

```python
df = pd.get_dummies(df, columns=['Precip Type'])
df.head()
```

```python
df = pd.get_dummies(df, columns=['Summary'])
df.head()
```

```python
df.shape
```

```python
columns_to_keep = ['Formatted Date', 'Temperature (C)', 'Apparent Temperature (C)', 'Humidity', 'Wind Speed (km/h)', 'Wind Bearing (degrees)',
        'Visibility (km)', 'Pressure (millibars)', 'Precip Type_rain', 'Precip Type_snow', 'Summary_Mostly Cloudy', 
        'Summary_Partly Cloudy', 'Summary_Overcast', 'Summary_Clear', 'Summary_Foggy']
df1 = df[columns_to_keep].copy()
df1.head()
```

```python
df1.shape
```

```python
df1.dtypes
```

```python
df1 = df1.replace({True: 1, False: 0})
df1.dtypes
```

```python
df1.head()
```

```python
df1 = df1.round(decimals=2)
df1.head()
```

```python
a = df1.drop('Formatted Date', axis=1)
korelacja = a.corr()
korelacja
```

```python
correlation_matrix = a.corr(method="spearman", numeric_only=True)
plt.figure(figsize=(12, 8))
plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='nearest')
plt.colorbar()
plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=60)
plt.yticks(range(len(correlation_matrix.columns)), correlation_matrix.columns)
plt.title('Correlation Matrix')
plt.show()
```

```python
df1 = df1.drop('Apparent Temperature (C)', axis=1)
df1.head()
```

```python
df1 = df1.drop('Wind Bearing (degrees)', axis=1)
df1.head()
```

```python
df1.shape
```

```python
df1_agg = df1.groupby('Formatted Date').agg(
                                        Visibility_mean = ("Visibility (km)", np.mean),
                                        Temperature_mean = ("Temperature (C)", np.mean),
                                        Humidity_mean = ("Humidity", np.mean),
                                        WindSpeed_mean = ("Wind Speed (km/h)", np.mean),
                                        Pressure_mean = ("Pressure (millibars)", np.mean))
df1_agg
```

```python
df1['Formatted Date'].value_counts()
```

```python
sns.set_theme(style="whitegrid")

values = df1_agg.drop('Pressure_mean', axis=1)
sns.lineplot(data=values, palette="tab10", linewidth=2.5)

# Jedynie 'Widoczno' w badanym okresie wykazuje trend wzrostowy (poprawia si). Takie zmienne jak 'Temperatura' i 'Prdko wiatru'
# na kracach badanego okresu maj podobne wartoci, jednak na uwag zasuguje okres 2009 - 2011, kiedy to nastpi stosunkowo
# gwatowny spadek odczytw, by nastpnie powrci do normy.
# Liczba pomiarw w 2006 roku: 8760, w 2016: 8410.

# Only 'Visibility' shows an upward trend (improving) in the analyzed period. Variables like 'Temperature' and 'Wind Speed'
# at the ends of the examined period have similar values, but the period 2009 - 2011 is worth attention, when there was 
# a relatively readings drop sharply and then return to normal.
# Number of measurements in 2006: 8760, in 2016: 8410.
```

```python
df2_agg = df1.groupby('Formatted Date').agg(
                                        Rain_mean = ("Precip Type_rain", np.sum),
                                        Snow_mean = ("Precip Type_snow", np.sum),
                                        MostlyCloudy_mean = ("Summary_Mostly Cloudy", np.sum),
                                        PartlyCloudy_mean = ("Summary_Partly Cloudy", np.sum),
                                        Overcast_mean = ("Summary_Overcast", np.sum),
                                        Clear_mean = ("Summary_Clear", np.sum),
                                        Foggy_mean = ("Summary_Foggy", np.sum))
df2_agg
```

```python
df1['Formatted Date'].value_counts() 
```

```python
values2 = df2_agg
sns.lineplot(data=values2, palette="tab10", linewidth=2.5)

# Pomimo wacha trudno okreli (na podstawie wartoci kracowych) zmiany trendw, z pewnym wyrnikiem dla 'MostlyCloudy' (wzrost),
# 'Snow' (spadek), 'Foggy' (spadek) czy 'Clear' (spadek). 
# Liczba pomiarw w 2006 roku: 8760, w 2016: 8410.

# Despite fluctuations.it is difficult to determine (based on marginal values) changes in trends, with a certain distinction
# for 'MostlyCloudy' (growth), 'Snow' (fall), 'Foggy' (fall) or 'Clear' (fall).
# Number of measurements in 2006: 8760, in 2016: 8410.
```
