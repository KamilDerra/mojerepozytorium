```python
a = pd.DataFrame({'EmployeeNumber': df4['EmployeeNumber'], 'YearsSinceLastPromotion': df4['YearsSinceLastPromotion'], 'YearsAtCompany': df4['YearsAtCompany'], 'YearsInCurrentRole': df4['YearsInCurrentRole']})
a
```

```python
def funkcja1(row):
    if row['YearsSinceLastPromotion'] > row['YearsAtCompany']:
        row['YearsSinceLastPromotion'] = row['YearsInCurrentRole']
    elif row['YearsInCurrentRole'] > row['YearsAtCompany']:
          row['YearsInCurrentRole'] = row['YearsSinceLastPromotion']
    return row

a = a.apply(funkcja1, axis=1)
a
```
