```python
ATTzeroAndATTone = pd.concat([ATTzero, ATTone], axis=0)
```

```python
from scipy.stats import ttest_ind
```

```python
# Szukamy kolumn z istotnymi różnicami (statystycznie) między histogramami ze zbiorów ATTzero i ATTone:
# (test t-Studenta)

# Wartość p, czyli p-wartość, to prawdopodobieństwo, że zależność jaką zaobserwowano w losowej próbie z populacji mogła
# wystąpić przypadkowo, wskutek losowej zmienności prób, choć w populacji wcale nie występuje. 
# W statystyce często stosuje się wartość alfa równą 0,05 jako wartość odcięcia dla znaczenia. 
# Jeśli wartość p jest mniejsza niż 0,05, odrzucamy hipotezę zerową, że nie ma różnicy między tymi środkami, 
# i stwierdzamy, że istnieje duża różnica .

# Są to potencjalne czynniki, na które warto zwrócić uwagę.

for column in ATTzeroAndATTone:
    statistic, pvalue = ttest_ind(ATTzero[column], ATTone[column])
    if pvalue < 0.05:
        print(column)
```
